# Plot Simulated Annealing Schedules

Simulated annealing is implemented as a variable factor that is multiplied to the target chemical potential.

### Import Modules

```python
import os
import sys

cwd = os.getcwd()
parent_dir = cwd + "/../.."
sys.path.insert(1, parent_dir)

print("Directory containing the notebook:")
print(cwd)
```

```python
from inspect import getmembers, isfunction

import numpy as np
import matplotlib.pyplot as plt

import analyses.adsorption.mu_schedules as ms
```

### Load Simulated Annealing Schedules

```python
schedules = [func[0] for func in getmembers(ms, isfunction)]
print("Simulated Annealing Schedules: ")
print(schedules)
```

### Plot Simulated Annealing Schedules

```python
N = 250                 # Number of snapshots
i = np.arange(0, N, 1)  # Snapshot array

for func_name in schedules:
    schedule = getattr(ms, func_name)
    y = [schedule(ind, N) for ind in i]

    plt.figure(figsize=(6, 3), dpi = 300)
    plt.plot(i, y)
    # plt.xlim(0, 200)
    # plt.ylim(-1.1, 1.1)
    plt.xlabel("Snapshot")
    plt.ylabel("Mu Adjustment Factor")
    plt.title(func_name.replace("_", " "))
    plt.show()
```
